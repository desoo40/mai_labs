```python
def my_func(x):
    return np.sin(x) + 2 * x * np.cos(x)**2 * np.sin(x)
```

```python
x = np.linspace(-15, 15, 10001)
y = my_func(x) #+ np.sin(x * 25.5)

plt.plot(x, y)
```

```python
from scipy.misc import derivative
```

```python
def grad(fun, x_0, alpha, eps):
    x = 1000
    x_l = list()
    x_l.append(x_0)
    while (1):
        if abs(x - x_0) < eps:
            break
        x = x_0
        x_0 = x - alpha * derivative(fun, x,  dx=1e-6)
        x_l.append(x_0)
    return x, np.array(x_l)
```

```python
def find_global(fun, alpha, eps, a, b, size):
    
    newArray = np.random.randint(a*m, b*m, size)/m
    m = fun(newArray[0])
    print(newArray)
    for el in newArray:
        x1, _ = grad(fun, el, alpha, eps)
        
        if (fun(x1) < m):
            m = x1
            
    return m
```
